### Grafica 2

```python
df = pd.DataFrame(columns = ['city'])

for document in res:
    for val in document["cityWinner"]:
        if val != None:
            if "y" in val and len(val) > 10:
                for city in val.split("y"):
                    df = df.append({'city' : city.strip()}, ignore_index = True)
            else:
                df = df.append({'city' : val}, ignore_index = True)

def getCoords(cities):
    lat = []
    long = []
    for city in cities:
        response = requests.get(f"http://open.mapquestapi.com/geocoding/v1/address?key=P7M7k3N8gceCqApBoq8LGuYPDFlXEMhx&location={city},Colombia").json()
        results = response["results"][0]['locations'][0]
        latitude = results["latLng"]["lat"]
        longitude = results["latLng"]["lng"]
        lat.append(latitude)
        long.append(longitude)
    return lat,long
```

```python
citiesList = df['city'].unique() #Lista de ciudades
df = df.groupby(df.columns.tolist(),as_index=False).size() # Lista de ciudad y su frencuencia
df["lat"], df["long"] = getCoords(df['city'])
```

```python
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app2 = JupyterDash(__name__, external_stylesheets=external_stylesheets)

# Create server variable with Flask server object for use with gunicorn
server = app.server

app.layout = html.Div([
    html.Div([dcc.Graph(id='balotos-bacanos2')])
])


@app.callback(
    dash.dependencies.Output('balotos-bacanos2', 'figure'),
    dash.dependencies.Input('balotos-bacanos2', 'id')
)
def make_graph(id):
    return px.density_mapbox(df, lat='lat', lon='long', z='size', radius=40,
                        center=dict(lat=5, lon=-75), zoom=4,
                        mapbox_style="stamen-terrain")
```
